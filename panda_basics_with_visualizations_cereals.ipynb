# Lets get some information.
__sort_values__ are used to sort the values of a file based on certain conditions.Let's sort top 10 cereals providing the protien

```python

top_protein=cdata.sort_values(['protein'], ascending=False)
top_protein.head(10)
```

# Visualizing the result

visulaizing the result using __matplotlib.pyplot__

```python
import matplotlib.pyplot as plt
%matplotlib inline
top_protein['protein'].head(10).plot(kind="barh")
plt.show()
```

# Using Pivot table
Pivot tabl helps to use a column as index and perform some operation on it<br/>
__name_of_panda_DataFrame.pivot_table(index=['col_name'])__

```python
sub_cdata=cdata[['name','carbo','fat','protein','vitamins']]
disp_by_name=sub_cdata.pivot_table(index=['name'])
```

```python
disp_by_name.head()
```

```python
disp_by_name.head().plot(kind='bar', figsize=(20,18))
plt.show()
```
